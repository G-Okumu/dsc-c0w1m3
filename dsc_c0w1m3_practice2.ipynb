# Practice for Measure of Dispersion

1. Load in the relevant dataset using csv DictReader. The path has been provided.

```python
# Your code here
import csv
path = "./nhis.csv"
```

2. Extract values for the the variable height and save to a Python list called `height_list`:

```python
# Your code here
with open(path) as dh:
    reader = csv.DictReader(dh)
    data = list(reader)
    

height_list = [int(x['height']) for x in data]
```

We have plotted a histogram of the data. There are clearly outliers in the data.

```python
# Run this cell without changes
import matplotlib.pyplot as plt
%matplotlib inline
# ^^This is a 'magic command' built into jupyter notebooks. We use it so that the visualization displays
# in the notebook directly, instead of in a separate window.
```

```python
# Run this cell without changes
# A histogram should display below
plt.hist(height_list, bins=8);
```

3.  Calculate the sample mean of the heights.

```python
# Your code here
import numpy as np

mean_value = np.mean(height_list)

mean_value
```

4. Calculate the sample variance of the heights. Make sure that you are using the corrected variance.

```python
# Your code here
height_variance = np.var( height_list,ddof=1)
print(height_variance)
```

5. Calculate the sample standard deviation of heights. Make sure that you are using the corrected standard deviation.

```python
# Your code here

stand_deviation = np.std(height_list, ddof=1)
stand_deviation
```

6. Calculate the IQR by calculating and subtracting relevant percentiles.

```python
# Your code here
q1 = np.percentile(height_list, 25)
q3 = np.percentile(height_list, 75)

iqr_list = q3 - q1

iqr_list
```

7. Calculate the IQR using scipy.

```python
# Your code here
from scipy.stats import iqr

iqr(height_list)
```

8. Clean all outliers from the height data. You can use the histogram of heights to help guide in your data filtering.

```python
# Your code here
cleaned_heights = [x for x in height_list if x <= 81]
cleaned_heights
```

9. Compute the sample standard deviation on the cleaned data:

```python
# Your code here
st_for_cleaned_data = np.std(cleaned_heights, ddof=1)
st_for_cleaned_data
```

10.  Compute the new IQR on the cleaned data:

```python
# Your code here
iqr(cleaned_heights)
```

11.  Comment on how the standard deviation and IQR has changed between  the original and cleaned heights.

The standard deviation has reduced substantially—to a value that now reflects a deviation from the mean about which the bulk of the main height distribution lies. The IQR has reduced somewhat but is basically stable. The IQR now is commensurate with the characteristic range 
 given by the sample standard deviation and mean.
